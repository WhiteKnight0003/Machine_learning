# Mối quan hệ giữa các features
image.png

- cộng tuyến : chỉ là mqh giữa các feature
- hệ số tương quan : mqh giữa các features và target

- Collinearity : cộng tuyến
- Multi collinearity : đa cộng tuyến

+ Mức độ 2 hay nhiều features quan hệ tuyến tính với nhau
+ Thông thường hệ số tương quan > 0.7 or < -0.7 biểu thị rằng 2 hay nhiều features có hiện tượng (đa) cộng tuyến với nhau


- ví dụ giữa 2 feature A và B ngta không mong hệ số tương quan nó sẽ cao vì khi hệ số tương quan giữa 2 feature cao sẽ rất ảnh hưởng đến model , các vấn đề có thể phát sinh ra do đa cộng tuyến or hệ số tương quan cao :
  + Ở cái hình trên bạn có thể thấy :    
      + bên trái là các vòng tròn (các feature - tác động đến target theo nhiều cách khác nhau - thứ chuung ta mong muốn)
      + bên phải : các feature lại chỉ ảnh hưởng đến target theo các cách tương đồng nhau
      + ví dụ với bệnh tiểu đường , cta mong chỉ số tiểu đường sẽ tác động đến bệnh 1 kiểu và máu sẽ tác động 1 kiểu để dễ xác nhận bệnh
      

- cách để xác định tầm qtrong của 1 biến :     
  + ví dụ có ptb1 : y = w1x1 + w2x2 + w3x3 + w4x4 + w0
  + ví dụ bạn muốn xem x1 có quuan trọng hay k - bạn sẽ cố định giá trị x2, x3, x4  - rồi tăng giảm x1
      + nếu x1 tăng giảm mà làm cho y tăng giảm nhiều => tức x1 rất quan trọng và ngược lại - nch là ta sẽ nhìn vào các giá trị w1, w2 , w3 , w4 để xem x1, x2 , x3, x4 có qtrong hay không

  + ví dụ nếu x1, x2 bị đa cộng tuyến với nhau (> 0.7 or < -0.7) lúc đó bạn sẽ rất khó để cố định x1 , x2 vì x1 giảm thì x2 cũng giảm - lúc này w1 sẽ không còn được công bằng nx =>> hay khi có đa cộng tuyến bạn sẽ không thể xác định được độ qtrong của các features

  + không phải lúc nào đa cộng tuyến cũng là vấn đề :
    + ví dụ khi mà bạn cần xác định tại sao 1 ng bị covid thì bạn buộc phải loại bỏ đa cộng tuyến (collinearity) - ví dụ những model mà nó yêu cầu bạn phải giải thích tại sao mà lại bị mắc covid
    + ví dụ khi dự đoán giá nhà , bạn cả qtam đến cái mẹ gì làm giá nhà tăng nên kệ mẹ nó (nch là đối với các model cần dự đoán càng đúng càng tốt - và không qtam đến việc bạn phải giải thích cho nó)
  
  + cách giảm thiểu đa cộng tuyến :  có 1 cách là trong 2 cột giữ 1 cột và bỏ 1 cột
