# Các phương pháp được thể hiện
1. Over-sampling (lấy mẫu tăng cường)
  - 1 số phương pháp :
      + RandomOverSampler : Tạo thêm các mẫu bằng cách sao chép ngẫu nhiên từ lớp thiểu số
      + Smote : Tạo các mẫu mới bằng cách nội suy giữa các điểm dữ liệu gần nhau
      + Adasyn : TƯơng tự smote nhưng tập trung tạo mẫu ở cùng khó phân loại hơn
2. Under-sampling (Lấy mẫy giảm thiểu)
  - sử dụng phương pháp ClusterCentroids:
      + Giảm số lượng mẫu của lớp đa số
      + Sử dụng clustering để chọn ra các điểm đại diện
3. So sánh hiệu quả
  - Over-Sampling
    + Giữ được nhiều thông tin của data
    + Có thể gây overfitting nếu đơn giản chỉ sao chép dữ liệu
    + smote và adasyn tạo ra dữ liệu mới có ý nghĩa hơn
  
  - Under-Sampling
    + Giảm kích thước tập dữ liệu , giúp huấn luyện nhan hơn
    + Có thể mất thông tin quan trọng từ lớp đa số
    + ClusterCentroids cố gắng giữ lại thông tin quan trọng thông qua clustering

4. Ứng dụng
  - Phù hợp cho các bài toán phân loại đối với dữ liệu mất cân bằng (imbalanced data)
    + Phát hiện gian lận
    + Chẩn đoán y tế
    + Phát hiện sự cố
    + Phân loại thư spam

```python
Khi ta dùng random over‐sampling (nhân đôi ngẫu nhiên các mẫu của lớp thiểu số) thì sẽ gặp một số vấn đề chính:

Overfitting (dữ liệu học máy “học thuộc lòng”)
– Vì ta chỉ sao chép y hệt các điểm đã có, mô hình rất dễ ghi nhớ chính xác những điểm này thay vì học được đặc trưng tổng quát của lớp thiểu số. 
Khi gặp dữ liệu mới hơi khác biệt, hiệu suất giảm mạnh.

Khuếch đại nhiễu (noise amplification)
– Bất kỳ “outlier” hoặc điểm nhiễu nào trong lớp thiểu số cũng sẽ bị nhân đôi, khiến mô hình hiểu sai vùng phân bố và dễ mắc lỗi hơn.

Phân bố sai lệch (distribution distortion)
– Việc lặp lại nguyên mẫu có thể tạo ra những vùng mật độ nhân tạo (clusters) không phản ánh phân bố thực của dữ liệu. 
Điều này có thể nới rộng biên phân lớp một cách không tự nhiên.

Không giải quyết được vấn đề vùng biên (boundary issue)
– Random over‐sampling không tạo mẫu mới ở biên giữa hai lớp, nên nếu dữ liệu gốc không đủ mẫu gần biên, mô hình vẫn thiếu 
thông tin để xác định rõ ranh giới.

So sánh với các phương pháp sinh mẫu tổng hợp (SMOTE, ADASYN)
SMOTE tạo mẫu mới bằng cách nội suy giữa các điểm lân cận, giảm overfitting so với việc nhân đôi, nhưng có thể sinh ra những điểm “lạc” nằm ở vùng quá gần biên, làm mờ ranh giới.

ADASYN tự động tập trung sinh thêm ở những vùng khó học (few‐shot), cải thiện khả năng phân biệt, nhưng cũng dễ tạo thêm nhiễu nếu các điểm gốc có noise.

Tóm lại, random over‐sampling rất dễ cài đặt nhưng thường làm mô hình bị overfitting và mất khả năng khái quát hóa. Nếu muốn cân bằng dữ liệu mà vẫn giữ được tính đa dạng, người ta thường ưu tiên các phương pháp sinh mẫu tổng hợp (SMOTE, ADASYN) hoặc kết hợp cả oversampling và undersampling một cách thông minh.
```
