```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('/content/gdrive/MyDrive/1_Data_Science_Machine_Learning/TaiLieu/datasets/diabetes.csv')
#data.hist() # biểu diễn bằng lược đồ histogram các cột = biểu đồ cột

data.plot(kind = 'density', subplots = True, layout =(3,3) , sharex = False )
# Đây là biểu đồ mật độ (Density Plot) hay còn gọi là biểu đồ ước lượng mật độ kernel (Kernel Density Estimation plot - KDE plot)
# kind='density': Chỉ định loại biểu đồ là density plot
# subplots=True: Tạo biểu đồ riêng cho từng cột trong dataset
# layout=(3,3): Sắp xếp các biểu đồ theo ma trận 3x3
# sharex=False: Mỗi biểu đồ có trục x riêng



plt.show()
```

```python
#Boxplot
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('/content/gdrive/MyDrive/1_Data_Science_Machine_Learning/TaiLieu/datasets/diabetes.csv')
data.plot(kind = 'box', subplots = True, layout =(3,3) , sharex = False )

plt.show()
```

```python
# Khi in báo cáo cho khách hàng hay đồng nghiệp bạn hãy dùng ma trận hệ số tương quan (Correlation matrix plot -  ma trận đối xứng qua đường chéo chính) để cho mọi ng dễ hình dung

# nhìn vào màu sẽ thấy màu càng sáng tương quan càng cao , màu càng tối tương quan càng thấp - đánh giá sự tương quuan giữa 2 feature khác nhau
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data = pd.read_csv('/content/gdrive/MyDrive/1_Data_Science_Machine_Learning/TaiLieu/datasets/diabetes.csv')
sns.heatmap(data.corr(), annot=True)
# data.corr() - truyền dữ liệu vào
# annot = True : hiện chi tiết dữ liệu
plt.show()
```

```python
# Scatter Matrix Plot

import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/gdrive/MyDrive/1_Data_Science_Machine_Learning/TaiLieu/datasets/diabetes.csv')
pd.plotting.scatter_matrix(data)
plt.show()
```
