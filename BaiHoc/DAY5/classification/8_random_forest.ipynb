Random Forest – Phân tích chuyên sâu

1. Giới thiệu cơ bản
- Random Forest là phương pháp ensemble dựa trên tập hợp nhiều cây quyết định (Decision Trees).
- Mỗi cây được huấn luyện độc lập trên một bootstrap sample từ tập huấn luyện.

2. Quy trình xây dựng mô hình
1. **Bootstrap Aggregation (Bagging)**:
   - Cho mỗi cây, lấy một mẫu ngẫu nhiên có hoàn lại (bootstrap) từ tập huấn luyện ban đầu.
2. **Xây dựng cây quyết định**:
   - Với mỗi nút phân tách, chỉ xem xét một **tập con ngẫu nhiên** của các feature (max_features) để tìm split tốt nhất.
3. **Lặp lại** cho đến khi xây đủ số cây (n_estimators) hoặc đạt điều kiện dừng.
4. **Kết hợp đầu ra**:
   - **Phân loại**: vote đa số (majority voting) từ các cây.
   - **Hồi quy**: lấy trung bình (averaging) dự đoán của các cây.

3. Ưu điểm chính
- Giảm overfitting so với cây đơn lẻ nhờ tính ngẫu nhiên và trung bình nhiều cây.
- Khả năng khái quát tốt, ít nhạy với nhiễu.
- Xử lý dữ liệu đa chiều, feature quan trọng có thể được đánh giá.

4. Nhược điểm
- Mất khả năng giải thích trực quan như cây đơn.
- Tốn tài nguyên tính toán và bộ nhớ (nhiều cây, nhiều cấu trúc).
- Có thể kém hiệu quả khi dữ liệu rất lớn nếu không điều chỉnh tham số.

5. Siêu tham số quan trọng
- **n_estimators**: số lượng cây trong rừng.
- **max_features**: số feature xem xét ở mỗi split (sqrt, log2, hoặc số nguyên).
- **max_depth**, **min_samples_split**, **min_samples_leaf**: điều khiển độ sâu và kích thước nút.
- **bootstrap**: có sử dụng bootstrap sampling hay không.
- **criterion**: tiêu chí phân tách (Gini hoặc Entropy cho phân loại; MSE, MAE cho hồi quy).

6. Tính năng đánh giá quan trọng (Feature Importance)
- Tính độ giảm impurity trung bình mà mỗi feature đóng góp (Mean Decrease in Impurity).
- Có thể dùng permutation importance để đánh giá tính ổn định.

7. Tăng tốc và tối ưu
- **Out-of-Bag (OOB) error**: đánh giá nội bộ không cần tập validation bằng cách dùng những mẫu không được bootstrap.
- **Parallel processing**: đào tạo song song các cây.
- **Giảm chiều**: loại bỏ feature ít quan trọng để tiết kiệm tài nguyên.

8. Ứng dụng điển hình
- Dự đoán giá bất động sản.
- Phân loại hình ảnh.
- Chẩn đoán y tế.
- Phân tích rủi ro tín dụng.
