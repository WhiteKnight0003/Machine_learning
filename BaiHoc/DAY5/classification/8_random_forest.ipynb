Random Forest  Phn tch chuyn su

1. Gii thiu c bn
- Random Forest l phng php ensemble da trn tp hp nhiu cy quyt nh (Decision Trees).
- Mi cy c hun luyn c lp trn mt bootstrap sample t tp hun luyn.

2. Quy trnh xy dng m hnh
1. **Bootstrap Aggregation (Bagging)**:
   - Cho mi cy, ly mt mu ngu nhin c hon li (bootstrap) t tp hun luyn ban u.
2. **Xy dng cy quyt nh**:
   - Vi mi nt phn tch, ch xem xt mt **tp con ngu nhin** ca cc feature (max_features)  tm split tt nht.
3. **Lp li** cho n khi xy  s cy (n_estimators) hoc t iu kin dng.
4. **Kt hp u ra**:
   - **Phn loi**: vote a s (majority voting) t cc cy.
   - **Hi quy**: ly trung bnh (averaging) d on ca cc cy.

3. u im chnh
- Gim overfitting so vi cy n l nh tnh ngu nhin v trung bnh nhiu cy.
- Kh nng khi qut tt, t nhy vi nhiu.
- X l d liu a chiu, feature quan trng c th c nh gi.

4. Nhc im
- Mt kh nng gii thch trc quan nh cy n.
- Tn ti nguyn tnh ton v b nh (nhiu cy, nhiu cu trc).
- C th km hiu qu khi d liu rt ln nu khng iu chnh tham s.

5. Siu tham s quan trng
- **n_estimators**: s lng cy trong rng.
- **max_features**: s feature xem xt  mi split (sqrt, log2, hoc s nguyn).
- **max_depth**, **min_samples_split**, **min_samples_leaf**: iu khin  su v kch thc nt.
- **bootstrap**: c s dng bootstrap sampling hay khng.
- **criterion**: tiu ch phn tch (Gini hoc Entropy cho phn loi; MSE, MAE cho hi quy).

6. Tnh nng nh gi quan trng (Feature Importance)
- Tnh  gim impurity trung bnh m mi feature ng gp (Mean Decrease in Impurity).
- C th dng permutation importance  nh gi tnh n nh.

7. Tng tc v ti u
- **Out-of-Bag (OOB) error**: nh gi ni b khng cn tp validation bng cch dng nhng mu khng c bootstrap.
- **Parallel processing**: o to song song cc cy.
- **Gim chiu**: loi b feature t quan trng  tit kim ti nguyn.
