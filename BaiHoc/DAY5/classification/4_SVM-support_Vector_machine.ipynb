Support Vector Machine (SVM) – Phân tích chuyên sâu

1. Ý tưởng cơ bản
- Mục tiêu: Tìm siêu phẳng (hyperplane) phân tách hai lớp sao cho khoảng cách (margin) giữa siêu phẳng và các điểm dữ liệu gần nhất (support vectors) được tối đa hóa.
- Trong 2D, siêu phẳng là đường thẳng; trong không gian nhiều chiều, là mặt phẳng tương ứng.

2. Hard-Margin vs. Soft-Margin
2.1 Hard-Margin SVM (Linearly Separable)
- Giả sử dữ liệu hai lớp hoàn toàn tách biệt.

2.2 Soft-Margin SVM (Không hoàn toàn tách tuyến tính)
- Cho phép vi phạm ranh giới với biến slack ξ_i.
- C cân bằng margin rộng và mức vi phạm.

1. Support Vectors và Margin
- Support Vectors: Điểm thỏa mãn y_i (wᵀ x_i + b) = 1 hoặc có slack > 0.
- Margin: Khoảng cách từ hyperplane đến support vectors, quyết định khả năng khái quát.

1. Kernel Trick (Xử lý phi tuyến)
- Ánh xạ φ: x → φ(x) để tách tuyến tính trong không gian cao hơn.
- Dùng kernel K(x_i, x_j) = φ(x_i)ᵀ φ(x_j) để tránh tính trực tiếp φ.
- Ví dụ:
  * Linear: K(x,z) = xᵀ z
  * Polynomial: K(x,z) = (γ xᵀ z + r)^d
  * RBF: K(x,z) = exp(-γ ||x - z||²)
  * Sigmoid: K(x,z) = tanh(γ xᵀ z + r)

1. Bài toán Đối ngẫu (Dual)
- maximize ∑ α_i - ½ ∑ α_i α_j y_i y_j K(x_i, x_j)
  sao cho 0 ≤ α_i ≤ C, ∑ α_i y_i = 0.
- α_i > 0 ⇔ x_i là support vector.
- w = ∑ α_i y_i φ(x_i), b từ support vector.

1. Siêu tham số chính
- C: điều chỉnh cứng/mềm của margin.
- Loại kernel và tham số (γ, degree, coef0).

1. Hàm quyết định & Dự đoán
- f(x) = ∑ α_i y_i K(x_i, x) + b
- ŷ = sign(f(x))

1. Ưu – Nhược điểm
Ưu điểm:
- Margin tối đa hóa giúp khái quát tốt.
- Kernel linh hoạt cho dữ liệu phi tuyến.
- Kháng nhiễu (soft-margin).
Nhược điểm:
- Tinh chỉnh kernel và tham số phức tạp.
- Tốn thời gian và bộ nhớ với tập lớn.
- Khó diễn giải khi phi tuyến.

1. Ứng dụng chính
- Nhận dạng chữ viết tay.
- Phân loại văn bản (spam, sentiment).
- Phân loại hình ảnh.
- Bioinformatics (gene/protein).
