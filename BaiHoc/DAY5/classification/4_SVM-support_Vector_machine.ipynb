2. Hard-Margin vs. Soft-Margin

2.2 Soft-Margin SVM (Khng hon ton tch tuyn tnh)
- Cho php vi phm ranh gii vi bin slack _i.

1. Support Vectors v Margin
- Support Vectors: im tha mn y_i (w x_i + b) = 1 hoc c slack > 0.
- Margin: Khong cch t hyperplane n support vectors, quyt nh kh nng khi qut.

1. Kernel Trick (X l phi tuyn)
- nh x : x  (x)  tch tuyn tnh trong khng gian cao hn.
- Dng kernel K(x_i, x_j) = (x_i) (x_j)  trnh tnh trc tip .
- V d:
  * Linear: K(x,z) = x z
  * Polynomial: K(x,z) = ( x z + r)^d
  * RBF: K(x,z) = exp(- ||x - z||)
  * Sigmoid: K(x,z) = tanh( x z + r)

1. Bi ton i ngu (Dual)
- maximize  _i -   _i _j y_i y_j K(x_i, x_j)
  sao cho 0  _i  C,  _i y_i = 0.
- _i > 0  x_i l support vector.
- w =  _i y_i (x_i), b t support vector.

1. Siu tham s chnh

- Loi kernel v tham s (, degree, coef0).

1. Hm quyt nh & D on
