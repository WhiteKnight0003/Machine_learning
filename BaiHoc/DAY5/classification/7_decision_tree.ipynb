Decision Tree – Phân tích chuyên sâu

1. Giới thiệu cơ bản
- Decision Tree là mô hình supervised learning dùng cho phân loại và hồi quy.
- Mô hình biểu diễn dưới dạng cây nhị phân: nút quyết định (decision nodes) và nút lá (leaf nodes).

2. Cấu trúc cây
- **Root Node** (nút gốc): điểm bắt đầu, chứa toàn bộ tập dữ liệu.
- **Decision Node** (nút quyết định): tại mỗi nút, chọn một feature và ngưỡng (threshold) để phân chia dữ liệu thành hai nhánh.
- **Leaf Node** (nút lá): chứa nhãn (classification) hoặc giá trị trung bình (regression) khi kết thúc phân tách.

3. Tiêu chí phân tách (Splitting Criteria)
- **Gini Impurity** (Phân loại):
  Gini = 1 – Σ (p_i)^2
  - p_i: tỉ lệ mẫu thuộc lớp i trong nút.
  - Chọn feature và ngưỡng sao cho giảm thiểu Gini tổng weighted.
- **Entropy (Information Gain)**:
  Entropy = – Σ p_i log2(p_i)
  - Information Gain = Entropy(parent) – Σ ( (n_k / n_parent) * Entropy(k) )
- **Mean Squared Error (MSE)** (Hồi quy):
  MSE = (1/n) Σ (y_i – ŷ)^2

4. Quá trình xây dựng cây
1. Từ nút gốc, tính tiêu chí phân tách cho mọi feature và ngưỡng.
2. Chọn phân tách tốt nhất (maximizes Information Gain hoặc minimizes Gini/MSE).
3. Tạo hai nút con và lặp lại quá trình với tập con cho đến khi:
   - Đạt độ sâu tối đa (max_depth).
   - Số mẫu tối thiểu tại nút (min_samples_split).
   - Không còn cải thiện đáng kể.

5. Pruning (Cắt tỉa)
- **Pre-Pruning**: dừng phân tách sớm dựa trên điều kiện như max_depth, min_samples_leaf.
- **Post-Pruning**: xây dựng cây đầy đủ rồi cắt giảm các nhánh không hiệu quả, bằng cách đánh giá trên tập validation (cost-complexity pruning).

6. Ưu – Nhược điểm
Ưu điểm:
- Dễ hiểu, trực quan (có thể trực tiếp diễn giải).
- Xử lý được cả dữ liệu số và phân loại.
- Không cần chuẩn hóa dữ liệu.
Nhược điểm:
- Dễ overfitting nếu không pruning.
- Cây không ổn định: dữ liệu thay đổi nhỏ có thể thay đổi cấu trúc cây.
- Khó mô hình hóa quan hệ phức tạp mà không tạo cây quá sâu.

7. Siêu tham số chính
- max_depth: độ sâu tối đa của cây.
- min_samples_split: số mẫu tối thiểu để phân tách.
- min_samples_leaf: số mẫu tối thiểu tại nút lá.
- max_features: số lượng feature được xem xét khi tìm phân tách.

8. Ứng dụng điển hình
- Hệ thống hỗ trợ quyết định y tế.
- Phân tích rủi ro tín dụng.
- Chấm điểm khách hàng, marketing phân khúc.
- Dự báo nhu cầu (với hồi quy).
