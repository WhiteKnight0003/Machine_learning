Decision Tree  Phn tch chuyn su

1. Gii thiu c bn
- Decision Tree l m hnh supervised learning dng cho phn loi v hi quy.
- M hnh biu din di dng cy nh phn: nt quyt nh (decision nodes) v nt l (leaf nodes).

2. Cu trc cy
- **Root Node** (nt gc): im bt u, cha ton b tp d liu.
- **Decision Node** (nt quyt nh): ti mi nt, chn mt feature v ngng (threshold)  phn chia d liu thnh hai nhnh.
- **Leaf Node** (nt l): cha nhn (classification) hoc gi tr trung bnh (regression) khi kt thc phn tch.

3. Tiu ch phn tch (Splitting Criteria)
- **Gini Impurity** (Phn loi):
  Gini = 1   (p_i)^2
  - p_i: t l mu thuc lp i trong nt.
  - Chn feature v ngng sao cho gim thiu Gini tng weighted.
- **Entropy (Information Gain)**:
  Entropy =   p_i log2(p_i)
  - Information Gain = Entropy(parent)   ( (n_k / n_parent) * Entropy(k) )
- **Mean Squared Error (MSE)** (Hi quy):
  MSE = (1/n)  (y_i  )^2

4. Qu trnh xy dng cy
1. T nt gc, tnh tiu ch phn tch cho mi feature v ngng.
2. Chn phn tch tt nht (maximizes Information Gain hoc minimizes Gini/MSE).
3. To hai nt con v lp li qu trnh vi tp con cho n khi:
   - t  su ti a (max_depth).
   - S mu ti thiu ti nt (min_samples_split).
   - Khng cn ci thin ng k.

5. Pruning (Ct ta)
- **Pre-Pruning**: dng phn tch sm da trn iu kin nh max_depth, min_samples_leaf.
- **Post-Pruning**: xy dng cy y  ri ct gim cc nhnh khng hiu qu, bng cch nh gi trn tp validation (cost-complexity pruning).

6. u  Nhc im
u im:
- D hiu, trc quan (c th trc tip din gii).
- X l c c d liu s v phn loi.
- Khng cn chun ha d liu.
Nhc im:
- D overfitting nu khng pruning.
- Cy khng n nh: d liu thay i nh c th thay i cu trc cy.
- Kh m hnh ha quan h phc tp m khng to cy qu su.

7. Siu tham s chnh
- max_depth:  su ti a ca cy.
- min_samples_split: s mu ti thiu  phn tch.
- min_samples_leaf: s mu ti thiu ti nt l.
- max_features: s lng feature c xem xt khi tm phn tch.
