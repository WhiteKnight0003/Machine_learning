K-Nearest Neighbours (K-NN)  Phn tch chuyn su

1.  tng c bn
- M hnh non-parametric: Khng cn hun luyn thc s, ch lu tr ton b d liu gc.
- Vi mt im mi, K-NN phn loi bng cch tham kho nhn ca k im d liu gn nht trong khng gian c trng.

2. C ch hot ng chi tit
   1. Chn k: s lng ln cn tham gia b phiu.
   2. nh ngha khong cch:
      - Euclidean: d(x, z) = sqrt( (x_j - z_j)^2).
      - Manhattan: d(x, z) =  |x_j - z_j|.
      - Minkowski, Chebyshev, hoc khong cch c trng s.
   3. Tnh khong cch t im mi x n mi x_i trong tp hun luyn.
   4. Chn k im c khong cch nh nht  tp N_k(x).
   5. B phiu (majority vote):
      - Phn loi: chn nhn xut hin nhiu nht.
      - (Tu chn) Weighted vote: trng s ngc khong cch.
   6. D on nhn cho x.

3. La chn k (siu tham s)
- k qu nh (v d k=1): d overfitting.
- k qu ln: d underfitting.
- Tm k tt nht qua cross-validation.

  1. Tin x l d liu
  - Chun ha (normalization/standardization)  trnh c trng bin  ln chi phi.
  - C th gim chiu (PCA, LDA)  gim curse of dimensionality.

  1. Ti u & Tng tc
  -  phc tp d on: O(nd).
  - Cu trc tm kim nhanh: KD-Tree, Ball-Tree.
  - Approximate NN (LSH, HNSW) cho d liu rt ln.

  1. u  Nhc im
  u im:
  - n gin, d trin khai.
  - Khng cn hun luyn phc tp.
  - Linh hot vi nhiu loi khong cch.
  - C th m rng cho hi quy.
  Nhc im:
  - Tn b nh (lu d liu gc).
  - D on chm nu n ln.
  - Km hiu qu khi chiu cao.
  - Cn tin x l cn thn.

  1. Bin th & ng dng
  - Weighted K-NN: im gn hn c trng s cao hn.
  - Distance-weighted regression cho hi quy.
  - ng dng: phn loi hnh nh nh, recommendation c bn, y sinh.
