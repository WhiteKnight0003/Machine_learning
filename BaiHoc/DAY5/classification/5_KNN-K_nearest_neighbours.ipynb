K-Nearest Neighbours (K-NN) – Phân tích chuyên sâu

1. Ý tưởng cơ bản
- Mô hình non-parametric: Không cần huấn luyện thực sự, chỉ lưu trữ toàn bộ dữ liệu gốc.
- Với một điểm mới, K-NN phân loại bằng cách tham khảo nhãn của k điểm dữ liệu gần nhất trong không gian đặc trưng.

2. Cơ chế hoạt động chi tiết
   1. Chọn k: số lượng lân cận tham gia bỏ phiếu.
   2. Định nghĩa khoảng cách:
      - Euclidean: d(x, z) = sqrt(Σ (x_j - z_j)^2).
      - Manhattan: d(x, z) = Σ |x_j - z_j|.
      - Minkowski, Chebyshev, hoặc khoảng cách có trọng số.
   3. Tính khoảng cách từ điểm mới x đến mọi x_i trong tập huấn luyện.
   4. Chọn k điểm có khoảng cách nhỏ nhất → tập N_k(x).
   5. Bỏ phiếu (majority vote):
      - Phân loại: chọn nhãn xuất hiện nhiều nhất.
      - (Tuỳ chọn) Weighted vote: trọng số ngược khoảng cách.
   6. Dự đoán nhãn cho x.

3. Lựa chọn k (siêu tham số)
- k quá nhỏ (ví dụ k=1): dễ overfitting.
- k quá lớn: dễ underfitting.
- Tìm k tốt nhất qua cross-validation.

  1. Tiền xử lý dữ liệu
  - Chuẩn hóa (normalization/standardization) để tránh đặc trưng biên độ lớn chi phối.
  - Có thể giảm chiều (PCA, LDA) để giảm curse of dimensionality.

  1. Tối ưu & Tăng tốc
  - Độ phức tạp dự đoán: O(n·d).
  - Cấu trúc tìm kiếm nhanh: KD-Tree, Ball-Tree.
  - Approximate NN (LSH, HNSW) cho dữ liệu rất lớn.

  1. Ưu – Nhược điểm
  Ưu điểm:
  - Đơn giản, dễ triển khai.
  - Không cần huấn luyện phức tạp.
  - Linh hoạt với nhiều loại khoảng cách.
  - Có thể mở rộng cho hồi quy.
  Nhược điểm:
  - Tốn bộ nhớ (lưu dữ liệu gốc).
  - Dự đoán chậm nếu n lớn.
  - Kém hiệu quả khi chiều cao.
  - Cần tiền xử lý cẩn thận.

  1. Biến thể & Ứng dụng
  - Weighted K-NN: điểm gần hơn có trọng số cao hơn.
  - Distance-weighted regression cho hồi quy.
  - Ứng dụng: phân loại hình ảnh nhỏ, recommendation cơ bản, y sinh.
