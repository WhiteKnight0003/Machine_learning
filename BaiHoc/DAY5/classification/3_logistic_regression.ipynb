```python
1. Logistic Regression
Khái niệm:
- Là mô hình hồi quy dùng để phân loại nhị phân (binary classification).
- Dự đoán xác suất p của một đầu ra nhị phân bằng hàm sigmoid:
    p = 1 / (1 + e^(-z)), với z = w^T x + b.

Cơ chế hoạt động:
1. Tính z = w^T x + b.
2. Áp dụng hàm sigmoid để ép giá trị đầu ra vào khoảng (0,1).
3. Sử dụng threshold (thường 0.5) để quyết định lớp nhãn.
4. Học tham số w, b bằng tối ưu hóa hàm log-loss qua gradient descent.

Ưu điểm:
- Dễ hiểu, dễ triển khai.
- Hiệu quả với tập dữ liệu lớn.
- Kết quả dưới dạng xác suất có thể diễn giải.

Nhược điểm:
- Chỉ phân loại nhị phân (mở rộng đa lớp cần chiến lược one-vs-rest hoặc softmax).
- Giả định tuyến tính giữa biến đầu vào và log-odds.

Ứng dụng:
- Chẩn đoán bệnh (bệnh/không bệnh).
- Phân loại tín dụng (mắc nợ/không mắc nợ).
```
