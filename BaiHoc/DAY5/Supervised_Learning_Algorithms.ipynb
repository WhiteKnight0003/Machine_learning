  - Linear Regression
  - Polynomial Regression

# Đối với Classification (phân loại) có 6 bài toán :     
  - Logistic Regression
  - Support Vector Machine
  - K-nearest neighbours
  - Navie Bayes
  - Decision Tree
  - Random Forset

- Trong ảnh - với những mô hình mà có dấu sao và màu xanh dương - tuy nó được dùng cho bài toán classification nhưng  nó vẫn sẽ có các phiên bản khác để dùng cho bài toán Regression

# A. REGRESSION (Hồi quy):

1. Linear Regression (Hồi quy tuyến tính):
  - Đặc điểm:
    + Mô hình đơn giản, dễ hiểu
    + Tìm mối quan hệ tuyến tính giữa biến đầu vào và đầu ra
  
  - Ưu điểm:
    + Dễ cài đặt và diễn giải
    + Tốc độ huấn luyện và dự đoán nhanh

  - Nhược điểm:
    + Chỉ mô hình hóa được quan hệ tuyến tính
    + Nhạy cảm với nhiễu và outliers

  - Nên dùng khi:
    + Dữ liệu có quan hệ tuyến tính
    + Cần mô hình dễ giải thích
    + Dự đoán giá nhà, doanh số bán hàng

2. Polynomial Regression (Hồi quy đa thức):
  - Đặc điểm:
    + Mở rộng của hồi quy tuyến tính
    + Có thể mô hình hóa quan hệ phi tuyến

  - Ưu điểm:
    + Linh hoạt hơn Linear Regression
    + Có thể fit được dữ liệu phức tạp

  - Nhược điểm:
    + Dễ bị overfitting nếu bậc đa thức cao
    + Tốn nhiều tính toán hơn

  - Nên dùng khi:
    + Dữ liệu có quan hệ phi tuyến
    + Cần độ chính xác cao hơn Linear Regression


B. CLASSIFICATION (Phân loại):

1. Logistic Regression:
  - Đặc điểm:
    + Phân loại nhị phân
    + Dựa trên xác suất

  - Ưu điểm:
    + Dễ cài đặt và diễn giải
    + Cho kết quả xác suất

  - Nhược điểm:
    + Chỉ phù hợp với dữ liệu phân tách tuyến tính

  - Nên dùng khi:
    + Phân loại nhị phân đơn giản
    + Cần biết xác suất phân loại

2. Support Vector Machine (SVM):
  - Đặc điểm:
    + Tìm siêu phẳng tối ưu phân tách các lớp
    + Có thể sử dụng kernel tricks

  - Ưu điểm:
    + Hiệu quả với dữ liệu có nhiều chiều
    + Robust với outliers

  - Nhược điểm:
    + Khó chọn kernel và tham số phù hợp
    + Chậm với dữ liệu lớn


  - Nên dùng khi:
    + Dữ liệu có nhiều features
    + Cần độ chính xác cao

3. K-Nearest Neighbors (KNN):
  - Đặc điểm:
    + Phân loại dựa trên khoảng cách
    + Không cần huấn luyện mô hình

  - Ưu điểm:
    + Đơn giản, dễ hiểu
    + Linh hoạt với dữ liệu phức tạp

  - Nhược điểm:
    + Chậm khi dự đoán
    + Nhạy cảm với nhiễu

  - Nên dùng khi:
    + Dữ liệu nhỏ, ít nhiễu
    + Cần mô hình đơn giản

4. Naive Bayes:
  - Đặc điểm:
    + Dựa trên định lý Bayes
    + Giả định độc lập giữa các features

  - Ưu điểm:
    + Nhanh và hiệu quả
    + Hoạt động tốt với dữ liệu text

  - Nhược điểm:
    + Giả định độc lập không thực tế

  - Nên dùng khi:
    + Phân loại văn bản, spam
    + Cần tốc độ xử lý nhanh

5. Decision Tree và Random Forest:
  - Đặc điểm:
    + Decision Tree: Mô hình dạng cây quyết định
    + Random Forest: Tập hợp nhiều Decision Trees

  - Ưu điểm:
    + Dễ hiểu và diễn giải
    + Xử lý được cả dữ liệu số và categorical

  - Nhược điểm:
    + Decision Tree dễ overfitting
    + Random Forest có thể chậm với dữ liệu lớn

  - Nên dùng khi:
    + Cần mô hình dễ giải thích
    + Dữ liệu có nhiều loại features khác nhau
