# Li ca machine learning n t u
# c 2 loi li :
  - Reducible Error (li c th gim thiu c)
    + Bias
    + Variance
  - Irreducible Error (li khng th gim thiu c )

# Bias c rt nhiu ngha trong AI v d :
- Sai lch gia prediction v actual label
- cho ta bit kh nng d on chnh xc ca model
- Bias cng b cng tt
- High bias means:
  + Overly-simplified model
  + Under-fiiting
  + High error on both training and test sets

image.png

# Variance (Trong AI n cng c rt nhiu ngha - v d :)
  + Mc  thay i ca  chnh xc ca prediction nu input thay i
  + cho ta bit kh nng tng qut ha ca model
  + Variance cng b cng tt
  + High variance means :     
      + Overly-complex model
      + Over-fitting (data n gin - model phc tp)
      + Low error on training set but high error on test set


# Bias v Variance u l cc li v bn u mun gim chng i - nhng bn khng th gim c 2 ci 1 lc m ch c th gim 1 trong 2 (gim ci ny th ci kia tng v ngc li)

-  y cc im  cng gn trung tm th cng chnh xc v ngc li

# ch nn tp trung vo 2 trng hp (1 ci  tt ch fix ci ti)
  + High Bias - Low variance
  + High variance - Low Bias
# 2 ci kia
  + Low Bias - Low variance - qu tt - sao phi ci thin ci tt - v trong thc t cng k gp u
  + High Bias - High Variance - qu t - ch c b v th ngay ci khc

# Phn bit Optimal Balance , Overfitting , Underfitting
- optimal Balance : trung ha c 2 yu t data v complex - ci m chng ta cn
- Overfitting : simple data , model complex
- Underfitting :  complex data , model simple


- Error = Variance + Bias^2
- ti sao error li c cng thc nh vy : ti chng k cng th nguyn variance l bc 2 cn bias l bc 1
- optimal model complexity : l ng  i qua v tr total Error thp nht (hay t li nht)
