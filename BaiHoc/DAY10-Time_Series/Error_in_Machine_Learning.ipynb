# Lỗi của machine learning đến từ đâu
# có 2 loại lỗi :
  - Reducible Error (lỗi có thể giảm thiểu được)
    + Bias
    + Variance
  - Irreducible Error (lỗi không thể giảm thiểu được )

# Bias có rất nhiều nghĩa trong AI ví dụ :
- Sai lệch giữa prediction và actual label
- cho ta biết khả năng dự đoán chính xác của model
- Bias càng bé càng tốt
- High bias means:
  + Overly-simplified model
  + Under-fiiting
  + High error on both training and test sets

image.png

# Variance (Trong AI nó cũng có rất nhiều nghĩa - ví dụ :)
  + Mức độ thay đổi của độ chính xác của prediction nếu input thay đổi
  + cho ta biết khả năng tổng quát hóa của model
  + Variance càng bé càng tốt
  + High variance means :     
      + Overly-complex model
      + Over-fitting (data đơn giản - model phức tạp)
      + Low error on training set but high error on test set


# Bias và Variance đều là các lỗi và bạn đều muốn giảm chúng đi - nhưng bạn không thể giảm cả 2 cái 1 lúc mà chỉ có thể giảm 1 trong 2 (giảm cái này thì cái kia tăng và ngược lại)

- Ở đây các điểm đỏ càng gần trung tâm thì càng chính xác và ngược lại

# chỉ nên tập trung vào 2 trường hợp (1 cái đã tốt chỉ fix cái tồi)
  + High Bias - Low variance
  + High variance - Low Bias
# 2 cái kia
  + Low Bias - Low variance - quá tốt - sao phải cải thiện cái tốt - và trong thực tế cũng k gặp đâu
  + High Bias - High Variance - quá tệ - chỉ có bỏ và thử ngay cái khác

# Phân biệt Optimal Balance , Overfitting , Underfitting
- optimal Balance : trung hòa cả 2 yếu tố data và complex - cái mà chúng ta cần
- Overfitting : simple data , model complex
- Underfitting :  complex data , model simple


- Error = Variance + Bias^2
- tại sao error lại có công thức như vậy : tại chúng k cùng thứ nguyên variance là bậc 2 còn bias là bậc 1
- optimal model complexity : là đường  đi qua vị trí total Error thấp nhất (hay ít lỗi nhất)
