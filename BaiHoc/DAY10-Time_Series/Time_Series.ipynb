# Recursive multi-step Time series Forecasting
- chỉ cần 1 model và từ 4 ngày trước dự đoán ra ngày thứ 5
- những ô xanh là những ô bạn buộc phải biết , ô vàng là ô dự đoán
- nhược điểm :
    + với những ô vàng nếu dự đoán sai thì các ô sau sẽ càng sai nhiều

# Direct multi-step Time series Forecasting
- Thay vì sử dụng 1 model - bạn sẽ sử dụng n model cho n là số giá trị trong tương lai mà bạn cần phải dự đoán (ví dụ : dự đoán thời tiết cho n ngày )
- Ưu điểm : luôn dùng những data có thật
- NHược điểm : tốn memory khi mà bạn phải chạy nhiều model
