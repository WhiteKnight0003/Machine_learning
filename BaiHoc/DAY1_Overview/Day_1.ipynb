- Supervised learning : hc c gim st
  + ch c cho hc cc data  c gn label
  + c nhiu ng dng v ng dng rng di nht
  + c 2 loi bi ton con trong n
    + classification - bi ton phn loi - c 1 tp data hu hn gi tr vo input v d on n xem n thuc class hay type m mnh  chun b
    + regression - bi ton hi quy - c 1 tp data v hn gi tr vo output


- Unsupervised learning : hc khng c gim st
  + ch c cho hc cc b unlabel data (d liu hon ton cha c nh nhn)
  , m hnh s t phi tm hiu cc ci tnh cht , c trng ca b d liu v t  a ra c 1 vi prediction  con ngi c th s dng
  + c 2 loi bi ton con :     
    + clustering : Bi ton phn nhm : phn chia cc nhm d liu thnh cc nhm nh da trn s tng ng d liu trn cc ci nhm con
    + Association : bi ton kt hp : dng  khm ph ra 1 quy lut no  t nhiu b d liu cho trc (hay recommendation system - h thng gi  - v d khi mua mn  A trn sp pi n s gi  mn  B , C cho bn - mn  B , C l nhng mn  m ngi khc  mua sau khi mua mn  A)

- vn cn 1 loi na gi l bn gim st - s kt hp ca 2 supervised v unsupervised

- Reinforcement learning : hc tng cng - ci kh nht - thng dnh cho hng nghin cu
  + nh trong cc cng ty tesla - h pht trin xe t li - xe ny s khng c th nghim  ngoi m s c s dng trong phng th nghim (mi trng m phng )-  y s c cc agent(tc nhn) - v AI s c m phng
  + v d khi xe i ng ng , dng n  n s nhn c nhng ci reward (phn thng ) , or khi n i sai hoc tng vo ngi khc n s nhn c nhng ci phn hi negative v t  n s ci thin hnh vi (cc phn hi ny l t mi trng )


- s khc nhau gia bi ton classification ca supervised learning v clustering ca unsupervised learning :    
  + v d bn mua  chi c hnh trn , tam gic , vung cho l tr (model) v gi bn mun chng chia lm 3 nhm
    + i vi classification :  l khi bn  ch r cho n u l hnh trn , hnh tam gic , ... n s chia thnh 3 nhm hnh dng khc nhau theo  mun ca bn
    + i vi clustering : l khi bn khng h ch cho n , nhng bn vn mun n chia ra 3 nhm : by gi thay v chia thnh cc hnh dng khc nhau , n c th chia thnh 3 nhm c mu sc khc nhau chng hn
