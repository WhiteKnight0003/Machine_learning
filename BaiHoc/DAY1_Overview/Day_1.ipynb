- Supervised learning : học có giám sát
  + chỉ được cho học các data đã được gán label
  + có nhiều ứng dụng và ứng dụng rộng dãi nhất
  + có 2 loại bài toán con trong nó
    + classification - bài toán phân loại - có 1 tập data hữu hạn giá trị vào input và dự đoán nó xem nó thuộc class hay type mà mình đã chuẩn bị
    + regression - bài toán hồi quy - có 1 tập data vô hạn giá trị vào output


- Unsupervised learning : học không có giám sát
  + chỉ được cho học các bộ unlabel data (dữ liệu hoàn toàn chưa được đánh nhãn)
  , mô hình sẽ tự phải tìm hiểu các cái tính chất , đặc trưng của bộ dữ liệu và từ đó đưa ra được 1 vài prediction để con người có thể sử dụng
  + có 2 loại bài toán con :     
    + clustering : Bài toán phân nhóm : phân chia các nhóm dữ liệu thành các nhóm nhỏ dựa trên sự tương đồng dữ liệu trên các cái nhóm con
    + Association : bài toán kết hợp : dùng để khám phá ra 1 quy luật nào đó từ nhiều bộ dữ liệu cho trước (hay recommendation system - hệ thống gợi ý - ví dụ khi mua món đồ A trên sọp pi nó sẽ gọi ý món đồ B , C cho bạn - món đồ B , C là những món đồ mà người khác đã mua sau khi mua món đồ A)

- vẫn còn 1 loại nữa gọi là bán giám sát - sự kết hợp của 2 supervised và unsupervised

- Reinforcement learning : học tăng cường - cái khó nhất - thường dành cho hướng nghiên cứu
  + như trong các công ty tesla - họ phát triển xe tự lái - xe này sẽ không được thử nghiệm ở ngoài mà sẽ được sử dụng trong phòng thí nghiệm (môi trường mô phỏng )- ở đây sẽ có các agent(tác nhân) - và AI sẽ được mô phỏng
  + ví dụ khi xe đi đúng đường , dừng đèn đỏ nó sẽ nhận được những cái reward (phần thưởng ) , or khi nó đi sai hoặc tông vào người khác nó sẽ nhận được những cái phản hồi negative và từ đó nó sẽ cải thiện hành vi (các phản hồi này là từ môi trường )


- sự khác nhau giữa bài toán classification của supervised learning và clustering của unsupervised learning :    
  + ví dụ bạn mua đồ chơi có hình tròn , tam giác , vuông cho lũ trẻ (model) và giờ bạn muốn chúng chia làm 3 nhóm
    + đối với classification :  là khi bạn đã chỉ rõ cho nó đâu là hình tròn , hình tam giác , ... nó sẽ chia thành 3 nhóm hình dạng khác nhau theo ý muốn của bạn
    + đối với clustering : là khi bạn không hề chỉ cho nó , nhưng bạn vẫn muốn nó chia ra 3 nhóm : bây giờ thay vì chia thành các hình dạng khác nhau , nó có thể chia thành 3 nhóm có màu sắc khác nhau chẳng hạn

# câu hỏi :
- ví dụ bạn có 1 mô hình được học dữ liệu với bộ label gồm 1 triệu con chó or 1 triệu con mèo - giờ dùng model đó để phân loại các bức ảnh khác

# trả lời :     
- luôn cần có bàn tay của con người để các label đầu ra là hoàn toàn chính xác - tại các model dù có chạy chính xác đến đâu vẫn sẽ có sai sót và việc áp dụng các model chỉ để việc đánh nhãn diễn ra nhanh hơn còn con người vẫn luôn phải kiểm tra lại
